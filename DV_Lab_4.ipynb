#Monthly changes in different food grain stocks


```python
df = pd.read_csv("/content/drive/MyDrive/DV /L4/Production_of_Agriculture_Stock-Ministry_of_Finance.csv")
df
```

```python
df = df.drop(["Frequency", "Unit"], axis=1)
df
```

```python
df.columns = ['Particulars', 'Sep-2013', 'Oct-2013', 'Nov-2013', 'Dec-2013', 'Jan-2014', 'Feb-2014','Mar-2014', 'Apr-2014', 'May-2014', 'Jun-2014']
df
```

```python
df.loc[df["Particulars"] == "Foodgrains Rice", "Particulars"] = "Rice"
df.loc[df["Particulars"] == "Foodgrains Unmilled Paddy in term of Rice", "Particulars"] = "Unmilled Paddy(in term of Rice)"
df.loc[df["Particulars"] == "Foodgrains Wheat", "Particulars"] = "Wheat"
df
```

```python
df.plot(kind='bar', figsize = (12,8))
x = [0, 1, 2, 3]
labels = ["Foodgrains", "Rice", "Unmilled Paddy(in term of Rice)", "Wheat"]
plt.xticks(x, labels, rotation ='vertical')
plt.xlabel("Foodgrains")
plt.ylabel("Stock")
```

```python
df.plot(kind='barh', figsize = (12,8))
y = [0, 1, 2, 3]
labels = ["Foodgrains", "Rice", "Unmilled Paddy (in term of Rice)", "Wheat"]
plt.yticks(y, labels, rotation ='horizontal')
plt.xlabel("Foodgrains")
plt.ylabel("Stock")
```

```python
df = df.drop(0)
df
```

```python
df_1 = pd.melt(df, id_vars=['Particulars'], var_name='Month', value_name='Stock')
df_1.head()
```

```python
f, ax = plt.subplots(2, 2, figsize = (18,12))

# Palattes list
colors = [
    'inferno',
    'cubehelix',
    'coolwarm',
    ['#ff0b0b', '#0b0bff', '#0bff0b', '#9300ff','#FF00FF', '#FFFF00', '#008000', '#800080', '#000080', '#800000']
]

# Titles list
headings = [
    'Continuous Color Palette(Inferno)',
    'Sequential Color Palette(cubehelix)',
    'Bipolar Sequential Color Palette(coolwarm)',
    'Custom Color Palette'
]

for title, color, ax in zip(headings, colors, ax.ravel()):
    sns.barplot(data=df_1, x='Particulars', y='Stock', hue='Month', ax=ax, palette=color)
    ax.set_title(title)
    ax.set_xlabel('Month')
    ax.set_ylabel('Stock (in Ton mn)')
    ax.legend(loc='upper left')

f.suptitle("Monthly changes in different food grain stocks -- Screen Display") 
plt.tight_layout()
plt.show()
```

```python
f, ax = plt.subplots(2, 2, figsize=(25,14))

# Palattes list
colors = [
    'inferno',
    'cubehelix',
    'coolwarm',
    ['#ff0000', '#0000ff', '#00ff00']
]

# Titles list
headings = [
    'Continuous Color Palette(Inferno)',
    'Sequential Color Palette(cubehelix)',
    'Bipolar Sequential Color Palette(coolwarm)',
    'Custom Color Palette'
]

for title, color, ax in zip(headings, colors, ax.ravel()):
    plot = sns.lineplot(x='Month',  y='Stock', data=df_1, ax=ax, hue = 'Particulars', marker='X', markersize=10, palette = color, legend = False)
    plot = sns.barplot(x='Month',  y='Stock', data=df_1, ax=ax, hue = 'Particulars', alpha=.8, palette = color)
    ax.set_title(title)
    
ax.legend(loc='upper left')
f.suptitle("Monthly changes in different food grain stocks -- Presentation")
plt.tight_layout()
plt.show()
```

#State wise Colleges of different Ayus Category


```python
df1 = pd.read_csv('/content/drive/MyDrive/DV /L4/MergerdAyush_Colleges_in_India_01-04-2021_0.csv')
df1.head()
```

```python
df1["Total Colleges"] = df1['No. of Colleges - Govt'] + df1['No. of Colleges - Non-Govt']
df1.head()
```

```python
df1 = df1.drop(["Admission Capacity - Govt", "Permitted Seats - Govt", "Admission Capacity - Non-Govt", "Permitted Seats - Non-Govt"], axis=1)
df1.head()
```

```python
df1.columns = ['State', 'Type', 'Govt Colleges', 'Non-Govt Colleges', 'Total Colleges']
df1.head()
```

```python
cross_tab_prop = pd.crosstab(index=df1['State'], columns=df1['Type'],
                             normalize="index")
cross_tab_prop.head()
```

```python
df1_1 = pd.pivot_table(df1, index=['State'], columns=['Type'], values='Total Colleges')
df1_1.head()
```

```python
df1_1 = df1_1.loc[df1_1.sum(axis=1).sort_values(ascending=True).index]
df1_1.head()
```

```python
df1_1.plot(kind='bar', stacked=True, colormap='tab10', figsize=(18, 12))

plt.legend(loc="upper left")
plt.xlabel("State")
plt.ylabel("Number of Colleges")

plt.show()
```

```python
df1_1.plot(kind='barh', stacked=True, colormap='tab10', figsize=(12, 10))

plt.legend(loc="lower right")
plt.xlabel("State")
plt.ylabel("Number of Colleges")

plt.show()
```

```python
ax = df1_1.plot.barh(align='center', stacked=True, figsize=(10, 6), colormap = "tab10")
plt.tight_layout()

title = plt.title('State wise Colleges of different Ayus Category -- Screen Display')

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(10)
        
legend = plt.legend(loc='lower right')
      
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.text(x+width/2, 
            y+height/2, 
            '{:.0f}'.format(width), 
            horizontalalignment='center', 
            verticalalignment='center',
            color='black',
            fontsize=6)
    
```

```python
f, ax = plt.subplots(2, 2, figsize = (18,12))

# Palattes list
colors = [
    'inferno',
    'cubehelix',
    'coolwarm',
    ['#9300ff', '#008000', '#800080', '#000080', '#800000']
]

# Titles list
headings = [
    'Continuous Color Palette(Inferno)',
    'Sequential Color Palette(cubehelix)',
    'Bipolar Sequential Color Palette(coolwarm)',
    'Custom Color Palette'
]
# for label in (ax.get_xticklabels() + ax.get_yticklabels()):
#     label.set_fontsize(10)
for title, palette, ax in zip(headings, colors, ax.ravel()):
    df1_1.plot(kind='barh', stacked=True, ax=ax, color = sns.color_palette(palette),align='center')

    for p in ax.patches:
        width, height = p.get_width(), p.get_height()
        x, y = p.get_xy() 
        ax.text(x+width/2, y+height/2, 
                '{:.0f}'.format(width), 
                horizontalalignment='center', 
                verticalalignment='center',
                color='white', fontsize=6)
    
    ax.set_title(title)
    ax.set_xlabel('State')
    ax.set_ylabel('Number of Colleges')
    ax.legend(loc='lower right')
    
f.suptitle("State wise Colleges of different Ayus Category -- Presentation") 
plt.tight_layout()
plt.show()
    
```
